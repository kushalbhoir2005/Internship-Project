```python
#Create New Columns
df['Trade_Balance'] = df['Export_Billion_USD'] - df['Import_Billion_USD']
```

```python
df['GDP_Per_Capita'] = (df['GDP_Billion_USD']*1000) / df['Population_Million']
```

```python
df['Economic_Stability'] = df['GDP_Growth_%'] - df['Inflation_Rate_%']
```

```python
plt.figure()
sns.histplot(df['GDP_Billion_USD'], kde=True)
plt.title("GDP Distribution")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='Country', y='GDP_Growth_%', data=df)
plt.title("GDP Growth by Country")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
