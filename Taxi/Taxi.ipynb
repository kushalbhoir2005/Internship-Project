```python
#Create New Columns
df['Fare_per_km'] = df['Total_Fare'] / df['Distance_km']
df['Fare_per_km']
```

```python
df['pickup_datetime'] = pd.to_datetime(df['Pickup_Date'] + " " + df['Pickup_Time'])
df['drop_datetime'] = pd.to_datetime(df['Drop_Date'] + " " + df['Drop_Time'])

df['Trip_Duration_Min'] = (df['drop_datetime'] - df['pickup_datetime']).dt.total_seconds()/60
df['Trip_Duration_Min'] 
```
